Get the district shapes for RJ: 

```python
df_dist = get_shape(scale = 'sub_district', state = 'RJ')

df_dist.head()
```

Get the city shapes for RJ: 

```python
df_muni = get_shape(scale = 'city', state = 'RJ')

df_muni.head()
```

Compare the aggregation levels: 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
_, ax = plt.subplots()

num_polygons = len(df_muni)
colors = np.random.rand(num_polygons, 3)  # RGB values between 0 and 1

df_muni.plot(ax=ax, color = colors)

df_dist.boundary.plot(ax=ax, color = 'black')

ax.axis('off')

ax.set_title('The colors represent different cities.\n The line represent different districts')

plt.show()
```
